## Data Cleaning
- Add new columns; total number of guests, and total nights spent
- select the columns to use
- rename the column names
- replace values in the status column

```python
data['guests'] = data['no_of_adults'] + data['no_of_children']
data['nights_spent'] = data['no_of_weekend_nights'] + data['no_of_week_nights']
```

```python
data = data[['guests', 'nights_spent','type_of_meal_plan', 'room_type_reserved', 'lead_time', 
      'arrival_year', 'arrival_month','market_segment_type', 'avg_price_per_room', 'booking_status']]
data.columns = ['guests', 'nights','meal_plan', 'room_type', 'lead_time', 
      'year', 'month','market', 'room_price', 'status']
```

```python
data['status'] = data['status'].str.replace('_',' ')
```
