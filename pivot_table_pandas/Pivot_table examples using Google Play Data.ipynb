Reset index does not work on column headers, only on row indices.

```python
df = pd.pivot_table(data, 
               index = 'category', 
               columns = 'type', 
               values = 'rating',
               aggfunc = ['mean','max'],
               fill_value=0).head()

df.reset_index()
```

### Removing the multi-index layers

#### Option 1. merge the levels

##### using map and join

```python
df = pd.pivot_table(data, 
               index = 'category', 
               columns = 'type', 
               values = ['rating','installs'],
               aggfunc = 'mean',
               fill_value=0).head()
df
```

```python
df.columns.map('_'.join)
```

```python
df.columns = df.columns.map('_'.join)
df
```

##### using list comprehension

**Example 1**

```python
df = pd.pivot_table(data, 
               index = 'category', 
               columns = 'type', 
               values = ['rating','installs'],
               aggfunc = 'mean',
               fill_value=0).head()
df.columns = [f'{i}_{j}' for i,j in df.columns]
df
```

**Example 2**

```python
df = pd.pivot_table(data [data['year'] > 2015], 
               index = 'category', 
               columns = ['type','year'], 
               values = ['rating','installs'],
               aggfunc = 'median',
               fill_value=0).head()
df
```

```python
df.columns = [f'{j} {i} {k}' for i,j,k in df.columns]
df
```

```python
df.reset_index()
```

#### Option 2: Drop some levels

```python
#droplevel article
df = pd.pivot_table(data, 
               index = 'year', 
               columns = 'type', 
               values = ['category'],
               aggfunc = ['count', 'first'],
               fill_value=0)
df.tail()
```

```python
df.columns = df.columns.droplevel(level = [0,1])
df.tail()
```

```python
df.reset_index()
```

##### Removing the 'column-level' name

```python
#delete the level name
df2 = df.rename_axis(columns=None)

#reset the index
df2.reset_index()
```

### 'dropna' parameter (Default: dropna = True)

```python
df = pd.DataFrame({
    'id': [1,2,3,4,5],
    'area_code': ['1111','1111','1112','1111','1112'],
    'agent': ['a','b','a','b','a'],
    'price': [62.5,65.5,23.5,59.3,26.5],
    'size': [3.5, 3.75, 2.1, 2.95, 1.65]
})
df.sort_values('area_code')
```

```python
df['area_code'] = pd.Categorical(
                        df['area_code'], 
                        categories=['1111', 
                                    '1112', 
                                    '1113'])
df['agent'] = pd.Categorical(
                        df['agent'], 
                        categories=['a', 'b', 'c'])
```

```python
pd.pivot_table(df,
              index = 'area_code',
              columns = 'agent',
              values = 'price',
              aggfunc = 'mean',
              dropna = False)
```

### 'margins' parameter (Default: margins = False)

```python
df = pd.DataFrame({
    'id': [1,2,3,4,5],
    'area_code': ['1111','1111','1112','1111','1112'],
    'agent': ['a','b','a','b','a'],
    'price': [62.5,65.5,23.5,59.3,26.5],
    'size': [3.5, 3.75, 2.1, 2.95, 1.65]
})
df['area_code'] = pd.Categorical(
                        df['area_code'], 
                        categories=['1111', 
                                    '1112', 
                                    '1113'])
df['agent'] = pd.Categorical(
                        df['agent'], 
                        categories=['a', 'b', 'c'])
df.sort_values('area_code')
```

```python
pd.pivot_table(df,
              index = 'area_code',
              columns = 'agent',
              values = 'price',
              aggfunc = 'sum',
              dropna = False,
              fill_value = 0,
              margins = True)
```

### 'margins_name' parameter (Default: margins_name='All')

```python
pd.pivot_table(df,
              index = 'area_code',
              columns = 'agent',
              values = 'price',
              aggfunc = 'sum',
              dropna = False,
              fill_value = 0,
              margins = True,
              margins_name = 'sum_total')
```
